```
from tensorflow.keras.preprocessing.text import one_hot
```

```
sent=[
      "the glass of the milk",
      "the glass of the juice",
      "the glass of the water",
      "I am a good boy",
      "I am a good developer",
      "your videos is good",
      "understand the meaning of word"
]
```

```
sent
```

```
#Vocabullary size
voc_size= 10000
```

**One hot representation**

```
onehot_rpr=[one_hot(words, voc_size) for words in sent]
print(onehot_rpr)
```

**Word embedding representation**

```
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```
import numpy as np
```

```
sent_length= 8
emdebed_docs= pad_sequences(onehot_rpr, padding='pre', maxlen= sent_length)
print(emdebed_docs)
```

```
model= Sequential()
dimension= 10
model.add(Embedding(voc_size, dimension, input_length= sent_length))
model.compile('adam', 'mse')
```

```
model.summary()
```

```
print(model.predict(emdebed_docs))
```

```
print(model.predict(emdebed_docs)[0])
```
